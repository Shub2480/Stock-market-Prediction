```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import yfinance as yf
from prophet import Prophet
```

```python
url = "https://ournifty.com/stock-list-in-nse-fo-futures-and-options.html#:~:text=NSE%20F%26O%20Stock%20List%3A%20%20%20%20SL,%20%201000%20%2052%20more%20rows%20"
r=requests.get(url,verify=False)
htmlcontent=r.content
soup=BeautifulSoup(htmlcontent,'html.parser')
    
```

```python
stock_list = [i.get_text().strip() for i in soup.find_all(attrs = {'class':['column-3']})]    
```

```python
for i in ['SYMBOL',
 'FINNIFTY',
 'MIDCPNIFTY',
 'BANKNIFTY',
 'NIFTY']:
    stock_list.remove(i)
```

```python
start='2020-08-01'
end='2025-12-31'
```

```python

final_df=pd.DataFrame(columns=['ds','y','yhat_lower','yhat_upper','yhat','Stock_name'])


for i in stock_list:
    name=i+'.NS'

    df_ = yf.download(name, start , end  )
    #df_.reset_index(drop=True,inplace=True)
    df_.reset_index(inplace=True)
    df_.drop(['Open','High','Low','Adj Close','Volume'],axis=1,inplace=True)
    df_.columns=['ds','y']
    model = Prophet()
    model.fit(df_)
    new_dates=model.make_future_dataframe(180,include_history=False)
    forecast=model.predict(new_dates)
    forecast=forecast[['ds','yhat_lower','yhat_upper','yhat']]
    new_df_=pd.concat([df_,forecast])
    new_df_['Stock_name']=i
    final_df=pd.concat([final_df,new_df_])
```

```python
pd.set_option('display.max_rows',3000)
```

```python
final_df.to_csv('Stock_timeseries_prediction.csv')
```
